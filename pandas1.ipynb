```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Name': ['Vicky','Bill'],'Age':[12,46]})
df2 = pd.DataFrame({'Name': ['John','Sabrina'],'Age':[37,25]})
df1
```

```python
df2
```

```python
#CONCAT HORIZONTALLY

pd.concat([df1, df2], axis=1)
```

```python
#CONCAT VERTICALLY

df3=pd.concat([df1,df2],axis=0)
df3
```

```python
df3=pd.concat([df1,df2])
df3
```

```python
df4=pd.DataFrame({'Name': ['Tyler', 'Natalie'], 'Age': [28,39], 'Hobby': ['Swimming', 'Reading']})
df4
```

```python
pd.concat([df3,df4], join = 'outer', ignore_index=True)
```

```python
pd.concat([df3,df4], join = 'inner', ignore_index=True)
```

```python
size=pd.DataFrame({'Colour': ['Red','Blue', 'Green'], 'Size': ['M','S','L']})
gender=pd.DataFrame({'Colour': ['Red','Blue', 'Yellow'], 'Sex':['Female', 'Female', 'Male']})
size
```

```python
gender
```

```python
#MERGE

pd.merge(size,gender, how='inner')
```

```python
pd.merge(size,gender, how='outer')
```

```python
pd.merge(size,gender, how='left')
```

```python
pd.merge(size,gender, how='right')
```

```python
left=pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0','A1','A2','A3'],
                     'B': ['B0','B1','B2','B3']})

right=pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                    'C': ['C0','C1','C2','C3'],
                    'D': ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='Key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left,right, on=['key1', 'key2'])
```

```python
pd.merge(left,right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left,right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left,right, how='left', on=['key1', 'key2'])
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
